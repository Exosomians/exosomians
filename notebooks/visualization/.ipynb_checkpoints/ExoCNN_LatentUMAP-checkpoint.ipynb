# ExoCNN latent UMAP plotter

```python
import os
os.chdir("../../")
os.environ["CUDA_VISIBLE_DEVICES"]="2"
```

```python
import exoNet
import scanpy as sc
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_name = "final"
```

```python
data = np.load(f"./Data/{data_name}/sequences.npy", allow_pickle=True)
data = data[:, :, :4]
data.shape
```

```python
ev_random_sequences = np.loadtxt("./Data/Random/ev.csv", delimiter=',', dtype='str')
ev_random_sequences.shape
```

```python
ic_random_sequences = np.loadtxt("./Data/Random/ic.csv", delimiter=',', dtype='str')
ic_random_sequences.shape
```

```python
random_sequences = np.concatenate([ic_random_sequences, ev_random_sequences], axis=0)
```

```python
char_encoder = {
    'N': -1,
    'A': 0, 
    'C': 1, 
    'G': 2,
    'U': 3
}
```

```python
random_sequences_encoded = exoNet.pep.seq_encoder(random_sequences, char_encoder, 50, unknown_char=True)
random_sequences_encoded.shape
```

```python
labels = np.load(f"./Data/{data_name}/labels.npy", allow_pickle=True)
labels = np.reshape(labels, (-1,))
labels.shape
```

```python
network = exoNet.models.DeepBind(seq_len=data.shape[1],
                                 n_channels=data.shape[2],
                                 n_classes=2,
                                 padding='valid',
                                 use_batchnorm=True,
                                 lr=0.00001,
                                 model_path=f"./models/DeepBind/{data_name}/",
                                 dropout_rate=0.1,
                                )
```

```python
network.restore_model_weights()
```

```python
predictions = network.model.predict(data)
```

```python
data_name = 'final'
raw_df = pd.read_csv(f"./Data/{data_name}/{data_name}.design.mat.subset.shortened.csv")
raw_df.index = raw_df['id']
raw_df.head()
```

```python
results = pd.DataFrame({'id': raw_df['id'].values, 'no': predictions[:, 0], 'yes': predictions[:, 1]})
results.head()
```

```python
ev_extreme_results = results[results['id'].str.contains('ev')]
ev_extreme_results = ev_extreme_results[ev_extreme_results['yes'] >= ev_extreme_results['yes'].quantile(.90)]
ev_extreme_results.shape
```

```python
ic_extreme_results = results[results['id'].str.contains('ic')]
ic_extreme_results = ic_extreme_results[ic_extreme_results['no'] >= ic_extreme_results['no'].quantile(.90)]
ic_extreme_results.shape
```

```python
results_df = pd.concat([ev_extreme_results, ic_extreme_results])
results_df.head()
```

```python
ev_extreme_sequences = raw_df.loc[ev_extreme_results['id'], 'seq']
ev_extreme_sequences.head()
```

```python
ic_extreme_sequences = raw_df.loc[ic_extreme_results['id'], 'seq']
ic_extreme_sequences.head()
```

```python
char_encoder = {
    'N': -1,
    'A': 0,
    'C': 1,
    'G': 2,
    'U': 3
}
```

```python
ev_extreme_sequences_encoded = exoNet.pep.seq_encoder(ev_extreme_sequences, char_encoder, 50, unknown_char=True)
ev_extreme_sequences_encoded.shape
```

```python
ic_extreme_sequences_encoded = exoNet.pep.seq_encoder(ic_extreme_sequences, char_encoder, 50, unknown_char=True)
ic_extreme_sequences_encoded.shape
```

```python
ev_extreme_labels = np.array(['YES'] * ev_extreme_sequences_encoded.shape[0])
ic_extreme_labels = np.array(['NO'] * ic_extreme_sequences_encoded.shape[0])
```

```python
extreme_sequences_labels = np.concatenate([ev_extreme_labels, ic_extreme_labels], axis=0)
extreme_sequences_labels.shape
```

```python
extreme_sequences_encoded = np.concatenate([ev_extreme_sequences_encoded, ic_extreme_sequences_encoded], axis=0)
extreme_sequences_encoded.shape
```

```python
extreme_latent_data = network.to_latent(extreme_sequences_encoded)
extreme_latent_data.shape
```

```python
ic_random_sequences_encoded = exoNet.pep.seq_encoder(ic_random_sequences, char_encoder, 50, unknown_char=True)
ic_random_sequences_encoded.shape
```

```python
ev_random_sequences_encoded = exoNet.pep.seq_encoder(ev_random_sequences, char_encoder, 50, unknown_char=True)
ev_random_sequences_encoded.shape
```

```python
ic_random_latent_data = network.to_latent(ic_random_sequences_encoded)
ic_random_latent_data
```

```python
ev_random_latent_data = network.to_latent(ev_random_sequences_encoded)
ev_random_latent_data
```

```python
ic_random_labels = np.array(['ic random'] * len(ic_random_latent_data))
ic_random_labels.shape
```

```python
ev_random_labels = np.array(['ev random'] * len(ev_random_latent_data))
ev_random_labels.shape
```

```python
random_labels = np.concatenate([ic_random_labels, ev_random_labels], axis=0)
```

```python
random_labels = np.reshape(random_labels, (-1,))
```

```python
random_latent_data = np.concatenate([ic_random_latent_data, ev_random_latent_data], axis=0)
random_latent_data.shape
```

```python
keep_idx = np.random.choice(random_labels.shape[0], 20000, replace=False)
```

```python
random_latent_data = random_latent_data[keep_idx, :]
random_labels = random_labels[keep_idx]
```

```python
sc.settings.figdir = "/media/pgdrive/sharif/exosomians/data/pdfs/UMAPs/"
sc.settings.set_figure_params(dpi=300)
```

```python
umap_data = np.concatenate([extreme_latent_data, random_latent_data], axis=0)
umap_labels = np.concatenate([extreme_sequences_labels, random_labels], axis=0)
umap_data.shape, umap_labels.shape
```

```python
exoNet.pl.plot_umap(umap_data, umap_labels, show=True, 
                    save=".deepbind.random.extreme.pdf")
```

```python
latent_data = network.to_latent(data)
latent_data.shape
```

```python
keep_idx = np.random.choice(latent_data.shape[0], 40000, replace=False)
```

```python
latent_adata = sc.AnnData(X=latent_data)
latent_adata.obs['label'] = labels
latent_adata
```

```python
latent_adata = latent_adata[keep_idx, :]
```

```python
umap_data = np.concatenate([latent_adata.X, random_latent_data], axis=0)
umap_labels = np.concatenate([latent_adata.obs['label'].values, random_labels], axis=0)
umap_data.shape, umap_labels.shape
```

```python
exoNet.pl.plot_umap(umap_data, umap_labels, show=True, 
                    save=".deepbind.all.random.pdf")
```
